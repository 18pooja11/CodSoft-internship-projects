```python
x_train = x_train.astype(int)
x_test = x_test.astype(int)
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.fit_transform(x_test)
```

```python
##linear regression
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train_scaled, y_train)
```

```python
y_pred= lr.predict(x_test_scaled)
```

```python
##plotting y_test and y_pred data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test, y_pred, c ='g')
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred, c ='red', label = 'predicted', marker = "x")
plt.scatter(y_test, y_pred, c ='black', label = 'True', marker = "o")
plt.xlabel('True values')
plt.ylabel('predicted values')
plt.title('True vs Predicted Values')
plt.legend()
plt.show()
```

```python
##checking performance of  regression model
```

```python
from sklearn.metrics import mean_squared_error
mse =mean_squared_error(y_test, y_pred)
mse
```

```python
from sklearn.metrics import mean_absolute_error
mae =mean_absolute_error(y_test, y_pred)
mae
```

# the lower value of mse and mae represents high acuracy of regression model
